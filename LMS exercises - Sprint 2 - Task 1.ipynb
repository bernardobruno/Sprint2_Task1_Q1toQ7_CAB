### Count of Animation movies (Hint: use the count() function)

```python
# Using a loop
genres=df["Genre"]
animation=0

for x in genres:
    if x=="Animation":
        animation+=1
print("The count of animation movies is: ",str(animation))
```

```python
# Using .count()
genres_count=df["Genre"].value_counts()
print(genres_count)
```

### Show the Comedy movies in the year 2007.

```python
year=df.groupby("Year")
year
year_2007=year.get_group(2007)
year_2007
year_2007[year_2007["Genre"]=="Comedy"]
```

### Count of Animation movies with more than 70% Audience Score.

```python
# Using a loop
audience_score=df.groupby("Audience score %")
audience_score
audience_score_over_70=audience_score.get_group(70)
audience_score_over_70

animation_over_70=0

for x in audience_score_over_70:
    if x in "Genres":
        animation_over_70+=1
print("The count of animation movies with more than 70% Audience score is: ",str(animation_over_70))
```
